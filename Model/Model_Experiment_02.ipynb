```python
#Crear un mapa de calor de las características numéricas con respecto a Diagnosis
#Conocer cuales características son más relevantes
from scipy.stats import kendalltau

# Calcula los coeficientes de correlación de Kendall
corr_matrix = pd.DataFrame(index=['Archivo', 'η1', 'η2', 'N1', 'N2', 'η', 'N', 'V', 'D', 'E', 'T', 'B',
       'CC'], columns=['refactor'])
for numeric_variable in corr_matrix.index:
    kendall_corr, _ = kendalltau(df[numeric_variable], df['refactor'])
    corr_matrix.at[numeric_variable, 'refactor'] = kendall_corr
# Convierte los coeficientes de correlación a números para el mapa de calor
corr_matrix = corr_matrix.astype(float)

# Aumenta el DPI al crear la figura
plt.figure(figsize=(8, 8), dpi=100)
# Crea un mapa de calor
sns.heatmap(corr_matrix.sort_values("refactor",ascending=False), annot=True, cmap="Blues", fmt=".2f", vmin=-1, vmax=1)
plt.title('Mapa de calor - Coeficientes de Correlación de Kendall')
plt.show()
```
